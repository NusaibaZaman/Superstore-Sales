```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales=pd.read_csv("/content/train.csv")
sales.head()
```

```python
sales.dropna(inplace=True)
```

```python
sales.describe()
```

Number of Orders in each State:

```python
plt.figure(figsize=(12, 8))
ax = sns.countplot(x = 'State',data = sales)

for bars in ax.containers:
    ax.bar_label(bars)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
```

Number of Orders in each Category:

```python
ax = sns.countplot(x ='Category',data=sales)

for bars in ax.containers:
    ax.bar_label(bars)
```

Number of Orders in each Sub-Category:

```python
plt.figure(figsize=(12, 8))
ax = sns.countplot(x ='Sub-Category',data = sales)

for bars in ax.containers:
    ax.bar_label(bars)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
plt.tight_layout()
```

Total Sales generated by each State:

```python
plt.figure(figsize=(12,10))

sns.barplot(x='State',y='Sales',data=sales,estimator=sum,ci=None)
plt.xticks(rotation=45,ha='right')

plt.xlabel('State')
plt.ylabel('Total Sales in amount')
plt.title('Total Sales generated by State')

plt.tight_layout()
plt.show()
```

Total Sales generated for each Category:

```python
plt.figure(figsize=(12,10))

sns.barplot(x='Category',y='Sales',data=sales,estimator=sum,ci=None)
plt.xticks(rotation=45,ha='right')

plt.xlabel('Category')
plt.ylabel('Total Sales in amount')
plt.title('Total Sales generated in relation with each Categories')

plt.tight_layout()
plt.show()
```

Total Sales in each Year:

```python
daily_sales = sales.groupby('Order Date')['Sales'].sum().reset_index()
daily_sales['Order Date'] = pd.to_datetime(daily_sales['Order Date'])
yearly_sales = daily_sales.resample('Y', on='Order Date').sum().reset_index()

plt.figure(figsize=(12, 8))
sns.lineplot(x='Order Date', y='Sales', data=yearly_sales)

plt.xlabel('Year')
plt.ylabel('Total Sales in Amount')
plt.title('Total Sales Per Year')
plt.show()
```
